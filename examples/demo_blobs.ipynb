```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from kmeans.kmeans import KMeansClustering
```

```python
X, _ = make_blobs(n_samples=200, n_features=2, centers=3, random_state=42)
```

```python
model = KMeansClustering(k=3)
labels = model.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(model.centroids[:,0], model.centroids[:,1],
            c=range(len(model.centroids)),
            marker="*", s=200)
```
